### WebScrapin de los jugadores de todos los equipos de Italia ###

```python
url = 'https://www.transfermarkt.es/inter-de-milan/kader/verein/46/saison_id/2023/plus/1'
headers = {'User-Agent': 'Mozilla/5.0'} #navegador sin bloqueos
response = requests.get(url, headers=headers)

if response.status_code == 200: #correcta
    soup = BeautifulSoup(response.content, 'html.parser')
    tabla = soup.find('table', class_='items') #se centra en la tabla
    jugadores = []

    #extraigo las filas solo con las columnas que me interesan
    filas = tabla.find_all('tr', class_=['odd', 'even'])
    for fila in filas:
        columnas = fila.find_all('td')
        if columnas:
            dorsal = columnas[0].text.strip()
            nombre = columnas[1].img['alt'].strip() if columnas[1].img else columnas[1].text.strip()
            edad = columnas[3].text.strip()
            altura = columnas[4].text.strip()
            pie = columnas[5].text.strip()
            valor_mercado = columnas[8].text.strip()

            jugadores.append({
                'Dorsal': dorsal,
                'Nombre': nombre,
                'Edad': edad,
                'Altura': altura,
                'Pie': pie,
                'Valor de Mercado': valor_mercado
            })

    df_inter = pd.DataFrame(jugadores)
    print(df_inter.head())
    df_inter.to_csv('jugadores_inter_milan.csv', index=False)
else:
    print(f'Error: {response.status_code}')
```

```python
#doy nombre al CSV
df3 = pd.read_csv('jugadores_italiana_completo.csv')

#eliminar las columnas que no necesito
df3 = df3.drop(columns=['Rk', 'Pos_y', 'Comp', 'Age_y', 'Born', '90s', 'npxG', 
                        'xAG', 'npxG+xAG', 'PrgR', 'Gls_90', 'Ast_90', 'G+A_90',
                        'G-PK_90', 'G+A-PK_90', 'xG_90', 'xAG_90', 'xG+xAG_90', 
                        'npxG_90','npxG+xAG_90'], errors='ignore')

#renombrar columnas
df3 = df3.rename(columns={'Player': 'Nombre', 'Pos_x': 'Posicion', 'Age_x': 'Edad', 
                          'Height': 'Altura', 'Nation': 'Origen', 'Squad': 'Equipo', 
                          'MP': 'Partidos_jugados', 'Starts': 'Titutlar', 'Min': 'Minutos', 
                          'Gls': 'Goles', 'Ast': 'Asistencias', 'PK': 'Goles_penalti', 
                          'PKatt': 'Penaltis_tirados', 'CrdY': 'Amarillas', 'CrdR': 'Rojas', 
                          'xG': 'Goles_esperadospor_ocasion', 'PrgC': 'Conducciones', 'PrgP': 'Pases'})

#guardar como nuevo CSV y Excel (para mi, verlo mas visual)
df3.to_csv('jugadores_italiana_completo_limpio.csv', index=False)
df3.to_excel('jugadores_italiana_completo_limpio.xlsx', index=False)
```

```python
#Eliminar duplicados
df3 = df3.drop_duplicates()

df3.shape
```

```python
#eliminar duplicados en nombre
df3 = df3.drop_duplicates(subset=["Nombre"])
df3.shape
```

```python
# Mostrar conteo de valores nulos por columna
print(df3.isnull().sum())
```

```python
# Reemplazar nulos en la columna 'Origen' por 'it ITA'
df3['Origen'] = df3['Origen'].fillna('it ITA')
```

```python
print(df3.isnull().sum())
```

```python
#Rellenar nulos en columnas numéricas con la media de cada una
cols_a_rellenar = ['Goles_esperadospor_ocasion', 'Conducciones', 'Pases']
for col in cols_a_rellenar:
    media = df3[col].mean()
    df3[col] = df3[col].fillna(media)
```

```python
print(df3.isnull().sum())
```

```python
#paso la altura de m a cm y lo convierto el float
df3["Altura_cm"] = (df3["Altura"].replace("-", pd.NA)
    .str.replace("m", "", regex=False).str.replace(",", ".", regex=False))

df3["Altura_cm"] = (pd.to_numeric(df3["Altura_cm"], errors="coerce")
    .mul(100).round().astype("Int64"))
```

```python
# Calcular % de goles y asistencias por 90 minutos
df3["GA_por_90"] = df3["G+A"] / (df3["Minutos"] / 90)

# Asegurarnos que Edad y Minutos son numéricos
df3["Edad"] = pd.to_numeric(df3["Edad"], errors="coerce")
df3["Minutos"] = pd.to_numeric(df3["Minutos"], errors="coerce")
```

```python
df3.to_csv('jugadores_italiana_completo_limpio.csv', index=False)
df3.to_excel('jugadores_italiana_completo_limpio.xlsx', index=False)
```
